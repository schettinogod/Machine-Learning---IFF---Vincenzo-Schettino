```python
for i, tweet in enumerate(df['tweet']):
    # Definir os emojis positivos que você deseja procurar
    emojis_positivos = [':)', '😃', '😄', '😊', '😁']

    # Verificar se algum emoji positivo está presente no tweet usando expressões regulares
    regex = '|'.join(re.escape(em) for em in emojis_positivos)
    possui_emoji_positivo = bool(re.search(regex, tweet))

    # Armazenar 1 se um emoji positivo está presente e 0 caso contrário na coluna "possui_emoji_positivo" para o tweet correspondente
    dataset.loc[i, 'Emoji-Positivo'] = 1 if possui_emoji_positivo else 0
```

```python
for i, tweet in enumerate(df['tweet']):
    # Definir os emojis negativos que você deseja procurar
    emojis_negativos = [':(', '😞', '😔', '🙁', '😢']

    # Verificar se algum emoji negativo está presente no tweet usando expressões regulares
    regex = '|'.join(re.escape(em) for em in emojis_negativos)
    possui_emoji_negativo = bool(re.search(regex, tweet))

    # Armazenar 1 se um emoji negativo está presente e 0 caso contrário na coluna "possui_emoji_negativo" para o tweet correspondente
    dataset.loc[i, 'Emoji-Negativo'] = 1 if possui_emoji_negativo else 0
```

```python
for i, tweet in enumerate(df['tweet']):
    # Verificar se a string "RT" está presente no tweet em letras maiúsculas ou minúsculas
    possui_rt = 'RT' in tweet.upper()

    # Armazenar 1 se um retweet está presente e 0 caso contrário na coluna "possui_rt" para o tweet correspondente
    dataset.loc[i, 'RT'] = 1 if possui_rt else 0
```

```python
for i, tweet in enumerate(df['tweet']):
    # Verificar se uma hashtag está presente no tweet usando expressão regular
    possui_hashtag = bool(re.search(r'#\w+', tweet))

    # Armazenar 1 se uma hashtag está presente e 0 caso contrário na coluna "possui_hashtag" para o tweet correspondente
    dataset.loc[i, '#'] = 1 if possui_hashtag else 0
```

```python
# Percorrer cada tweet no DataFrame
for i, tweet in enumerate(df['tweet']):
    # Verificar se uma URL está presente no tweet usando expressão regular
    possui_url = bool(re.search('https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+', tweet))

    # Armazenar 1 se uma URL está presente e 0 caso contrário na coluna "possui_url" para o tweet correspondente
    dataset.loc[i, 'URL'] = 1 if possui_url else 0
```

```python
df["tweets"] = df["tweet"].str.lower()
# Percorrer cada tweet no DataFrame
for i, tweet in enumerate(df['tweets']):
    # Converter todas as letras para minúsculas para evitar a diferenciação entre letras maiúsculas e minúsculas
    tweet = tweet.lower()

    # Inicializar um dicionário para rastrear a contagem de cada letra
    contagem_letras = {}

    # Percorrer cada letra no tweet
    for letra in tweet:
        # Verificar se a letra é um caractere alfabético
        if letra.isalpha():
            # Verificar se a letra já existe no dicionário de contagem
            if letra in contagem_letras:
                # Incrementar a contagem da letra existente
                contagem_letras[letra] += 1
            else:
                # Inicializar a contagem da letra
                contagem_letras[letra] = 1

    # Calcular a soma das ocorrências das letras repetidas
    soma_letras_repetidas = sum(contagem for contagem in contagem_letras.values() if contagem > 1)

    # Armazenar a soma das letras repetidas na coluna "soma_letras_repetidas" para o tweet correspondente
    dataset.loc[i, 'LETRAS-REPETIDAS'] = soma_letras_repetidas
```

```python
# Percorrer cada tweet no DataFrame
for i, tweet in enumerate(df['tweet']):
    # Dividir o tweet em palavras
    palavras = tweet.split()

    # Contar o número de palavras no tweet
    num_palavras = len(palavras)

    # Armazenar o número de palavras na coluna "num_palavras" para o tweet correspondente
    dataset.loc[i, 'NUM-PALAVRAS'] = num_palavras
```

```python
for i, tweet in enumerate(df['tweet']):
    # Encontrar todas as palavras que iniciam com letra maiúscula usando expressão regular
    palavras_iniciais_maiusculas = re.findall(r'\b[A-Z]\w*\b', tweet)

    # Contar o número de palavras iniciais maiúsculas no tweet
    num_palavras_iniciais_maiusculas = len(palavras_iniciais_maiusculas)

    # Armazenar o número de palavras iniciais maiúsculas na coluna "num_palavras_iniciais_maiusculas" para o tweet correspondente
    dataset.loc[i, 'Palav-Ini-Maiusculas'] = num_palavras_iniciais_maiusculas
```

```python
for i, tweet in enumerate(df['tweet']):
    # Encontrar todas as palavras maiúsculas usando expressão regular
    palavras_maiusculas = re.findall(r'\b[A-Z]+\b', tweet)

    # Contar o número de palavras maiúsculas no tweet
    num_palavras_maiusculas = len(palavras_maiusculas)

    # Armazenar o número de palavras maiúsculas na coluna "num_palavras_maiusculas" para o tweet correspondente
    dataset.loc[i, 'Palavras-Maiusculas'] = num_palavras_maiusculas
```

```python
for i, tweet in enumerate(df['tweet']):
    # Contar o número de letras maiúsculas no tweet
    num_letras_maiusculas = sum(1 for letra in tweet if letra.isupper())

    # Armazenar o número de letras maiúsculas na coluna "num_letras_maiusculas" para o tweet correspondente
    dataset.loc[i, 'Letras-Maiusculas'] = num_letras_maiusculas
```

```python
for i, tweet in enumerate(df['tweet']):
    # Contar o número de pontos de exclamação no tweet
    num_pontos_exclamacao = tweet.count('!')

    # Armazenar o número de pontos de exclamação na coluna "num_pontos_exclamacao" para o tweet correspondente
    dataset.loc[i, '!-Quantidade'] = num_pontos_exclamacao
```

```python
for i, tweet in enumerate(df['tweet']):
    # Contar o número de pontos de interrogação no tweet
    num_pontos_interrogacao = tweet.count('?')

    # Armazenar o número de pontos de interrogação na coluna "num_pontos_interrogacao" para o tweet correspondente
    dataset.loc[i, '?-Quantidade'] = num_pontos_interrogacao
```

```python
for i, tweet in enumerate(df['tweet']):
    # Dividir o tweet em palavras separadas por espaço
    palavras = tweet.split()

    # Calcular o número total de caracteres e o número total de palavras
    num_caracteres_total = sum(len(palavra) for palavra in palavras)
    num_palavras = len(palavras)

    # Calcular a média de caracteres por palavra
    if num_palavras > 0:
        media_caracteres_por_palavra = num_caracteres_total / num_palavras
    else:
        media_caracteres_por_palavra = 0.0

    # Armazenar a média de caracteres por palavra na coluna "media_caracteres_por_palavra" para o tweet correspondente
    dataset.loc[i, 'Media-Caracteres'] = media_caracteres_por_palavra
```

```python
import string

punct_list = "\"$%&'()*+,-./:;<=>@[\]^_`{|}~?!"
exclist = punct_list + string.digits
# remove punctuations and digits
table = str.maketrans('', '', exclist)

def standardize_tweets(dataframe, attribute):
  dataframe[attribute] = dataframe[attribute].str.lower()
  dataframe[attribute] = dataframe[attribute].str.replace(r"@\S+","user")
  dataframe[attribute] = dataframe[attribute].str.replace(r"http\S+","url")
  dataframe[attribute] = dataframe[attribute].str.replace(r"home office","")
  dataframe[attribute] = dataframe[attribute].str.replace(r"home ofice","")
  dataframe[attribute] = dataframe[attribute].str.replace(r"trabalho remoto","")
  dataframe[attribute] = dataframe[attribute].str.translate(table) #retira pontuação e dígitos

standardize_tweets(df,"tweet")
```
